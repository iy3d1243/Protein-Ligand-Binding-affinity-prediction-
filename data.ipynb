```python
import pandas as pd
import numpy as np
import os

def remove_high_null_columns_chunked(input_file, output_file, threshold=0.5, chunksize=10000):
    """
    Remove columns from a large CSV file where more than threshold% of values are null/NaN.
    Uses chunked processing to handle files that don't fit in memory.
    
    Parameters:
    - input_file: path to input CSV/TSV file
    - output_file: path to save cleaned CSV file
    - threshold: fraction of null values above which to remove column (default 0.5 for 50%)
    - chunksize: number of rows to process at a time (default 10000)
    """
    
    print("🔍 Analyzing large dataset in chunks...")
    print(f"📁 Input file: {input_file}")
    print(f"📊 Processing {chunksize:,} rows at a time")
    
    # First pass: Count nulls and total rows for each column
    null_counts = {}
    total_rows = 0
    column_names = None
    
    try:
        # Use iterator to process file in chunks
        chunk_iterator = pd.read_csv(input_file, chunksize=chunksize, sep='\t' if input_file.endswith('.tsv') else ',')
        
        print("\n📈 First pass: Counting null values...")
        chunk_num = 0
        
        for chunk in chunk_iterator:
            chunk_num += 1
            
            # Initialize column tracking on first chunk
            if column_names is None:
                column_names = chunk.columns.tolist()
                null_counts = {col: 0 for col in column_names}
                print(f"📋 Found {len(column_names)} columns")
            
            # Count nulls in this chunk
            chunk_nulls = chunk.isnull().sum()
            for col in column_names:
                null_counts[col] += chunk_nulls.get(col, 0)
            
            total_rows += len(chunk)
            
            # Progress update
            if chunk_num % 10 == 0:
                print(f"   Processed {chunk_num} chunks ({total_rows:,} rows so far)...")
        
        print(f"✅ Analysis complete: {total_rows:,} total rows processed")
        
    except Exception as e:
        print(f"❌ Error reading file: {e}")
        return None
    
    # Calculate null percentages and determine which columns to keep
    print(f"\n🧮 Calculating null percentages...")
    null_percentages = {col: null_counts[col] / total_rows for col in column_names}
    
    columns_to_keep = [col for col in column_names if null_percentages[col] <= threshold]
    columns_to_remove = [col for col in column_names if null_percentages[col] > threshold]
    
    print(f"\n📊 Results:")
    print(f"   Original columns: {len(column_names)}")
    print(f"   Columns to remove: {len(columns_to_remove)} (>{threshold*100}% null)")
    print(f"   Columns to keep: {len(columns_to_keep)}")
    
    # Show worst offenders
    print(f"\n🗑️  Columns being removed (showing worst 10):")
    sorted_removed = sorted([(col, null_percentages[col]) for col in columns_to_remove], 
                           key=lambda x: x[1], reverse=True)
    for col, pct in sorted_removed[:10]:
        print(f"   - {col}: {pct*100:.1f}% null")
    if len(sorted_removed) > 10:
        print(f"   ... and {len(sorted_removed)-10} more")
    
    # Second pass: Write cleaned data
    print(f"\n💾 Second pass: Writing cleaned dataset...")
    
    try:
        chunk_iterator = pd.read_csv(input_file, chunksize=chunksize, sep='\t' if input_file.endswith('.tsv') else ',')
        
        first_chunk = True
        chunk_num = 0
        
        for chunk in chunk_iterator:
            chunk_num += 1
            
            # Keep only desired columns
            chunk_cleaned = chunk[columns_to_keep]
            
            # Write to file (header only on first chunk)
            chunk_cleaned.to_csv(output_file, 
                                mode='a' if not first_chunk else 'w',
                                header=first_chunk,
                                index=False)
            first_chunk = False
            
            # Progress update
            if chunk_num % 10 == 0:
                print(f"   Written {chunk_num} chunks...")
        
        print(f"✅ Cleaned dataset saved to: {output_file}")
        
    except Exception as e:
        print(f"❌ Error writing cleaned file: {e}")
        return None
    
    # Final summary
    print(f"\n📈 Final Summary:")
    print(f"   Input file: {input_file}")
    print(f"   Output file: {output_file}")
    print(f"   Rows processed: {total_rows:,}")
    print(f"   Original columns: {len(column_names)}")
    print(f"   Final columns: {len(columns_to_keep)}")
    print(f"   Reduction: {len(columns_to_remove)} columns removed ({len(columns_to_remove)/len(column_names)*100:.1f}%)")
    
    return columns_to_keep

# Memory-efficient version for very large files
def get_file_info(input_file):
    """Get basic info about the file without loading it entirely"""
    try:
        # Read just the header
        sample = pd.read_csv(input_file, nrows=0, sep='\t' if input_file.endswith('.tsv') else ',')
        
        # Get file size
        file_size_mb = os.path.getsize(input_file) / (1024 * 1024)
        
        print(f"📋 File Info:")
        print(f"   File size: {file_size_mb:.1f} MB")
        print(f"   Number of columns: {len(sample.columns)}")
        print(f"   Column names preview: {list(sample.columns[:5])}...")
        
        return len(sample.columns), file_size_mb
        
    except Exception as e:
        print(f"❌ Error reading file info: {e}")
        return None, None

# Usage
if __name__ == "__main__":
    # Your file paths
    input_file = "C:\\Users\\iyed\\Downloads\\BindingDB_All_202509_tsv\\BindingDB_All.tsv"  # Your input file
    output_file = "C:\\Users\\iyed\\Downloads\\BindingDB_All_202509_tsv\\BindingDB_All2.tsv"    # Your output file
    
    # Check file info first
    print("🔍 Checking file information...")
    num_cols, file_size = get_file_info(input_file)
    
    if num_cols and file_size:
        
        chunksize = 10000
        
        print(f"📊 Using chunk size: {chunksize:,} rows")
        
        # Process the file
        kept_columns = remove_high_null_columns_chunked(
            input_file=input_file,
            output_file=output_file,
            threshold=0.5,  # Remove columns with >50% nulls
            chunksize=chunksize
        )
        
        if kept_columns:
            print("\n🎉 SUCCESS! Dataset cleaned and saved.")
        else:
            print("\n❌ FAILED! Check error messages above.")
    else:
        print("❌ Could not analyze input file. Check file path and format.")
```
